```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv("/Users/narkhann/Research_Work/semester-2/deployment_of_ml_models/mlops_classroom/datasets/Iris.csv")
```

```python
df.sample(4)
```

```python
# unique flowers
df["Species"].unique()
```

#### Splitting data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
features = df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalLengthCm"]]
target = df["Species"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(features,target,test_size=0.5,shuffle=True)
```

```python
# Step 1 
model = DecisionTreeClassifier(criterion="entropy")
```

```python
model = DecisionTreeClassifier(criterion="entropy", max_depth=7, min_samples_leaf=4, random_state=42)

"""
    Hyperparameters - max_depth, min_samples_leaf
    
"""
```

```python
# Step 2 
model.fit(x_train,y_train)
```

```python
# Step 3 Predict it on unknown sample
unknown_data = [[5.5,4.6,1,2.4]]
model.predict(unknown_data)
```

```python
# step 4 Performance Evaluation
from sklearn.metrics import accuracy_score
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
pd.DataFrame({"actual": y_test, "predicted": y_pred})
```
